```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.size'] = 16
import numpy as np
!mkdir figures
```

```python
months = ['Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct']
fig, axs = plt.subplots(1, 2, figsize=(5*1.5*2,3*1.5))
ax, ax1 = axs
x = np.linspace(0, 13)
ax.plot(x, np.sin((x+3)*np.pi/12*2), lw=3)
ax_1 = ax.twinx()
ax_1.plot(x, -np.sin((x+3)*np.pi/12*2), lw=3, color='orange', linestyle='--')
ax_1.set_ylabel("Deformation", fontsize=18)
ax.set_ylabel("Mass of water, ice, snow", fontsize=18)
ax.hlines(0, 12, 0, color='k')
ylim = ax.get_ylim()
ax.fill_betweenx(ylim, 1, 6, color='skyblue', alpha=0.5, label='Wet season')
plt.xlim(0, 12)
ax.set_xticks(np.arange(13))
ax.set_xticklabels(months)
ax.set_ylim(ylim)
# ax.legend()
ax.set_yticks([-0.5, 0, 0.5])
ax_1.set_yticks([-0.5, 0, 0.5])
ax.set_yticklabels(["-", "0", "+"])
ax_1.set_yticklabels(["-", "0", "+"])

ax.yaxis.label.set_color('C1')
ax_1.yaxis.label.set_color('C0')


ax1.plot(x, -np.sin((x+3-0.3)*np.pi/12*2), lw=3)
ax1_1 = ax1.twinx()
ax1_1.plot(x, -np.sin((x+3)*np.pi/12*2), lw=3, color='orange', linestyle='--')
ax1_1.set_ylabel("Deformation", fontsize=18)
ax1.set_ylabel("Hydraulic Head", fontsize=18)
ax1.hlines(0, 12, 0, color='k')
ylim = ax1.get_ylim()
ax1.fill_betweenx(ylim, 1, 6, color='skyblue', alpha=0.5, label='Wet season')
plt.xlim(0, 12)
ax1.set_xticks(np.arange(13))
ax1.set_xticklabels(months)
ax1.set_ylim(ylim)
# ax1.legend()
ax1.set_yticks([-0.5, 0, 0.5])
ax1_1.set_yticks([-0.5, 0, 0.5])
ax1.set_yticklabels(["-", "0", "+"])
ax1_1.set_yticklabels(["-", "0", "+"])

ax1.yaxis.label.set_color('C1')
ax1_1.yaxis.label.set_color('C0')

ax.text(2.2, 0.8, 'Wet season')
ax.text(8, 0.8, 'Dry season')
ax.set_title("(a) Seasonal loading response")
ax1.set_title("(a) Seasonal poroelastic response")
fig.autofmt_xdate() 
plt.tight_layout()
fig.savefig("./figures/figure_s1", dpi=200)
```
